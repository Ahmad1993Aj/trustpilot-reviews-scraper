```python
# preprcessing:
umlaut_map = {
    'ä': 'ae',
    'ö': 'oe',
    'ü': 'ue',
    'Ä': 'Ae',
    'Ö': 'Oe',
    'Ü': 'Ue',
    'ß': 'ss'}

for k, v in umlaut_map.items():
    df['text'] = df['text'].str.replace(k, v, regex=False)
    df['title'] = df['title'].str.replace(k, v, regex=False)

df["text"] = df["text"].str.strip()
df["title"] = df["title"].str.strip()
```

```python
month_map = {
    'Januar': '01',
    'Februar': '02',
    'März': '03',
    'April': '04',
    'Mai': '05',
    'Juni': '06',
    'Juli': '07',
    'August': '08',
    'September': '09',
    'Oktober': '10',
    'November': '11',
    'Dezember': '12'
}
```

```python
def parse_german_date(date_str)-> str:
    if not isinstance(date_str, str):
        return None
    parts = date_str.replace(".", "").split()
    if len(parts) != 3:
        return None
    day, month_name, year = parts
    month_num = month_map.get(month_name)
    if not month_num:
        return None
    return f"{year}-{month_num}-{day.zfill(2)}"
```

```python
df["date"] = df["date"].apply(parse_german_date)
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python
df.drop("date_parsed", axis=1, inplace=True, errors="ignore")
```

```python
# emoji durch text ersetzen:
import emoji

def preprocess_emoji(text: str) -> str:
    if not isinstance(text, str):
        return ""
    text = emoji.demojize(text, delimiters=(" ", " "))
    text = text.replace("_", " ")
    return " ".join(text.split())


df["text"] = df["text"].apply(preprocess_emoji)
df["title"] = df["title"].apply(preprocess_emoji)
```
